# Axis Roster Information 

An Axis Roster must have exactly $53$ players in total.

Rosters must first have a total of $30$ starters IN THE FOLLOWING ORDER:
* QB 1
* RBs 1-2
* WRs 1-5
* TE 1
* LT 1
* LG 1 
* C 1
* RG 1
* RT 1
* LE 1, DTs 1-2, RE 1 (Total of 4 DLs; distribution may vary depending on team)
* LOLB 1, MLBs 1-2, ROLB 1 (Total of 4 LBs; distribution may vary depending on team)
* LCB 1, LCB 2, SS 1, FS 1, RCB 1, RCB 2 (Total of 6 DBs; distribution may vary depending on team)
* K 1 
* P 1
Rosters must then have $23$ unused players as backups.


In the code cell below, we run GenerateAxisRoster all $32$ teams, which updates the axis roster for each team. 

```python
teams=[i for i in os.listdir('Team Mods') if i !='.DS_Store']
for team in teams:
    try:
        GenerateAxisRoster(team)
    except:
        continue
```
